# Bài 1: Viết code để các đèn led R,G,B lần lượt sáng dần rồi tối dần. Thời gian sáng dần và tối dần là 2s.

```python
#include <stdio.h>
#include <wiringPi.h>
#include <softPwm.h>
#define led1_pwm 18

#define led2_pwm 16
#define led3_pwm 11
int duty = 0 ;
int A[3]= {led1_pwm ,led2_pwm ,led3_pwm };
int main (void)
{
  wiringPiSetupPhys();
  //tao pwm
  softPwmCreate (led1_pwm,0,100 );
  softPwmCreate (led2_pwm,0,100 );
  softPwmCreate (led3_pwm,0,100 );


  /////////////////////////////
    printf("raspberrypi \n");
    
  // pinMode(led1_pwm,    PWM_OUTPUT);
  //  pinMode(led2,OUTPUT);  
    //pinMode(led3,OUTPUT); 

    while(1)
    {
        for (int i = 0 ; i <3; i++)
        {
            for (duty= 0;duty<= 100 ; duty+=10)
              {
                softPwmWrite(A[i],duty);
                delay(500);
              }
              softPwmWrite(A[i],0);
        } 
        
       // digitalWrite(led2,HIGH);
       // digitalWrite(led3,HIGH);
           
    }
}


```

# Bài 2: Viết code để 4 nút nhấn điều khiển độ sáng của đèn, với BT1 -> sáng yếu nhất, BT4 sáng nhất.

```python
#include <stdio.h>
#include <wiringPi.h>
#include <softPwm.h>
#define led1_pwm 18
#define bt1 37
#define bt2 36
#define bt3 38
#define bt4 40
//#define led2_pwm 16
//#define led3_pwm 11
int duty ;

int main (void)
{
  wiringPiSetupPhys();
  //tao pwm
softPwmCreate (led1_pwm,0,100 );
pinMode (bt1, INPUT) ;
pinMode (bt2, INPUT) ;
pinMode (bt3, INPUT) ;
pinMode (bt4, INPUT );
 // softPwmCreate (led2_pwm,0,100 );
  //softPwmCreate (led3_pwm,0,100 )

  /////////////////////////////
    printf("raspberrypi \n");
    
  // pinMode(led1_pwm,    PWM_OUTPUT);
  //  pinMode(led2,OUTPUT);  
    //pinMode(led3,OUTPUT); 

    while(1)
    {
        if( digitalRead(bt1))
        {
            duty =0;
             softPwmWrite(led1_pwm ,duty);   
        }
        
          else if( digitalRead(bt2))
        {
            duty =20;
             softPwmWrite(led1_pwm ,duty);   
        }
        
              else if( digitalRead(bt3))
        {
            duty =40;
             softPwmWrite(led1_pwm ,duty);   
        }
        
              else if( digitalRead(bt4))
        {
            duty =80;
             softPwmWrite(led1_pwm ,duty);   
        }
        
        

       // digitalWrite(led2,HIGH);
      //  digitalWrite(led2,HIGH);
       // digitalWrite(led3,HIGH);
           
    }
}


```
